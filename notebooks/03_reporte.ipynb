# Informe Forense — Análisis de Incidente de Seguridad



**Proyecto:** Security Log Analyzer
**Dataset:** Sherlock "Brutus" — HackTheBox
**Fuente:** /var/log/auth.log
**Analista:** Daniel Reyes

## Resumen Ejecutivo




El día 6 de marzo, el servidor `ip-172-31-35-28` sufrió un ataque de fuerza
bruta automatizado contra el servicio SSH, originado desde la dirección IP
**65.2.161.68**. El atacante logró comprometer la cuenta `root` en menos de
10 segundos desde el inicio del ataque, estableció persistencia mediante la
creación de una cuenta de usuario no autorizada, y procedió a exfiltrar
credenciales del sistema.



### Hallazgos clave

| Indicador | Detalle |
|---|---|
| IP atacante | 65.2.161.68 |
| Vector de ataque | Fuerza bruta SSH + enumeración de usuarios |
| Intentos fallidos totales | 48 |
| Duración del ataque inicial | 9 segundos |
| Cuenta comprometida | root |
| Hora de la brecha | 06:31:40 |
| Usuario backdoor creado | cyberjunkie |
| Privilegios otorgados | Grupo sudo |
| Datos comprometidos | /etc/shadow (hashes de contraseñas) |
| Persistencia instalada | Script linper.sh descargado vía curl |

### Cronología del ataque



1. **06:31:31** — Inicio de enumeración de usuarios (admin, backup, server_adm, svc_account)
2. **06:31:37** — Pivote hacia ataque de fuerza bruta dirigido a `root`
3. **06:31:40** — Acceso exitoso como `root`
4. **06:34:18** — Creación del usuario `cyberjunkie` (UID 1002)
5. **06:35:15** — Usuario `cyberjunkie` añadido al grupo `sudo`
6. **06:37:34** — Login SSH exitoso como `cyberjunkie`
7. **06:37:57** — Lectura de `/etc/shadow` vía sudo
8. **06:39:38** — Descarga de script de persistencia externo

### Severidad: CRÍTICA



El atacante obtuvo acceso completo de root, exfiltró credenciales del
sistema e instaló mecanismos de persistencia. El compromiso es total.
